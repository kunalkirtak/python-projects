# **Order Model**

```python
"""
==========================================================
Order Model using Pydantic
==========================================================
"""

from pydantic import BaseModel, Field, EmailStr


class Customer(BaseModel):

    customer_id: int

    name: str

    email: EmailStr


class Product(BaseModel):

    product_id: int

    name: str

    price: float = Field(gt=0)


class Order(BaseModel):

    order_id: int

    customer: Customer

    product: Product

    quantity: int = Field(gt=0)

    @property
    def total_amount(self):

        return self.product.price * self.quantity


customer = Customer(
    customer_id=1,
    name="Kunal",
    email="kunal@example.com"
)

product = Product(
    product_id=1001,
    name="Laptop",
    price=65000
)

order = Order(
    order_id=501,
    customer=customer,
    product=product,
    quantity=2
)

print("=" * 60)
print("ORDER MODEL")
print("=" * 60)

print(order)

print("\nOrder Total : ₹", order.total_amount)

print("\nSerialized Order")

print(order.model_dump())

print("\nOrder JSON")

print(order.model_dump_json(indent=4))
```
