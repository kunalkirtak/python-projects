# **Password Checker**

```python
# ==========================================
# Password Strength Checker
# ==========================================

print("=" * 45)
print("Password Strength Checker")
print("=" * 45)

password = input("Enter your password: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

special_characters = "!@#$%^&*()-_=+[]{};:'\",.<>?/\\|`~"

for char in password:
    if char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True
    elif char.isdigit():
        has_digit = True
    elif char in special_characters:
        has_special = True

print("\nPassword Analysis")
print("-" * 30)

if len(password) >= 8:
    print("✓ Minimum Length")
else:
    print("✗ Minimum Length (8 characters required)")

print("✓ Uppercase Present" if has_upper else "✗ Uppercase Missing")
print("✓ Lowercase Present" if has_lower else "✗ Lowercase Missing")
print("✓ Digit Present" if has_digit else "✗ Digit Missing")
print("✓ Special Character Present" if has_special else "✗ Special Character Missing")

if len(password) >= 8 and has_upper and has_lower and has_digit and has_special:
    print("\nPassword Strength : Strong")
elif len(password) >= 6:
    print("\nPassword Strength : Medium")
else:
    print("\nPassword Strength : Weak")
```
