# **Age Calculator**

```python
# ==========================================
# Age Calculator
# ==========================================

print("=" * 40)
print("Age Calculator")
print("=" * 40)

name = input("Enter your name: ")

birth_year = int(input("Enter your birth year: "))

current_year = int(input("Enter current year: "))

age = current_year - birth_year

print("\n----- Result -----")
print(f"Name : {name}")
print(f"Age  : {age} Years")
```
